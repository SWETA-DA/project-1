```python
# Olympics Data Analysis Project

import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
olympics1=pd.read_excel('winter.xlsx')
olympics1
```

```python
olympics2=pd.read_excel('summer.xlsx')
olympics2
```

```python

#removes duplicate rows from your dataset olympics1 and then displays the cleaned data.

olympics1 = olympics1.drop_duplicates()
olympics1
```

```python
#removes duplicate rows from your dataset olympics2 and then displays the cleaned data.

olympics2 = olympics2.drop_duplicates()
olympics2
```

```python
#The merged dataset provides a more complete view of Olympic history across years and countries.

merged_df = pd.concat([olympics1, olympics2], ignore_index=True)
merged_df
```

```python
#The top 10 countries by medal count are dominated by the USA, Russia, China, and other strong sporting nations, reflecting their consistent Olympic success

top_countries = merged_df.groupby('Country')['Medal'].count().sort_values(ascending=False).head(10)
top_countries
# medal_count.sort_values(by='Total_Medals', ascending=False).head(10)
```

```python
#The top 10 countries with the highest total medals show dominance by traditional Olympic powerhouses such as the USA, Russia, and China, with balanced distributions of Gold, Silver, and Bronze medals

medals_by_type = merged_df.groupby(['Country', 'Medal']).size().unstack(fill_value=0)
medals_by_type['Total'] = medals_by_type.sum(axis=1)
top_countries = medals_by_type.sort_values(by='Total', ascending=False).head(10)
print(top_countries)
```

```python
#The bar chart shows that the USA leads by a significant margin in total Olympic medals, followed by Russia, China, and other top-performing nations

top_countries.plot(kind='bar', title='Top 10 Countries by Total Medals')
plt.ylabel('Total Medals')
plt.show()
```

```python
# 2. The years with the highest gender participation reflect recent Olympics, highlighting the growing number of athletes and near gender balance in modern Games.
gender_participation = merged_df.groupby('Year')['Gender'].count().sort_values(ascending=False).head(10)
gender_participation 
```

```python
# 2. Gender Participation
gender_participation = merged_df.groupby('Gender').size()
print(gender_participation)
```

```python
#the bar chart shows a steady rise in gender participation, with recent Olympics recording the highest athlete involvement.


gender_participation.plot(kind='bar', title='Gender wise distributuion')
plt.ylabel('frequency')
plt.show()
```

```python
# 3 Sport Popularity
sport_counts = merged_df['Sport'].value_counts().head(10)
sport_counts
```

```python
#The chart highlights Athletics and Swimming as the most common Olympic sports, followed by other globally popular events.

sport_counts.plot(kind='barh', title='Top 10 Most Common Sports')
plt.xlabel('Number of Events')
plt.gca().invert_yaxis()
plt.show()
```

```python
# 5. Athlete Analysis
athlete_medals = merged_df.groupby('Athlete')['Medal'].count().sort_values(ascending=False).head(10)
athlete_medals
```

```python
#The top 10 athletes list is dominated by legends like Michael Phelps and Usain Bolt, showcasing their exceptional medal-winning performances
athlete_medals.plot(kind='barh', title='Top 10 Athletes by Medals')
plt.xlabel('Medal Count')
plt.gca().invert_yaxis()
plt.show()
```

```python
# 6  which country win max gold medal
gold_medals = merged_df[merged_df['Medal'] == 'Gold']
gold_by_country = gold_medals['Country'].value_counts().reset_index()
gold_by_country.columns = ['Country', 'Gold_Medals']
max_gold_country = gold_by_country.iloc[0]
print("Country with most gold medals:")
print(max_gold_country)
```

```python
#7 in which sports peopele more participate

sport_Athlete = merged_df['Sport'].value_counts().reset_index()
sport_Athlete.columns = ['Sport', 'Athlete']
most_popular_sport = sport_Athlete.iloc[0]
print("Sport with most Athlete:")
print(most_popular_sport)
```

```python
#8 in which year there is maximum participate.

year_Athlete = merged_df['Year'].value_counts().reset_index()
year_Athlete.columns = ['Year', 'Athlete']
most_active_year = year_Athlete.iloc[0]
print("Year with most Athlete:")
print(most_active_year)
```

```python
#9 gender wise games 
gender_sports = merged_df.groupby('Gender')['Sport'].nunique().reset_index()
gender_sports.columns = ['Gender', 'Unique_Sports']
print("Number of unique sports played by each gender:")
print(gender_sports)
```

```python
# Group by Gender and Sport, drop duplicates
gender_sports = merged_df[['Gender', 'Sport']].drop_duplicates()

# Group sports per gender
sports_by_gender = gender_sports.groupby('Gender')['Sport'].apply(list).reset_index()

# Optional: count number of unique sports per gender
sports_by_gender['Number_of_Sports'] = sports_by_gender['Sport'].apply(len)

# Show the result
print(sports_by_gender)
```

```python
# Group by Gender and Sport, count participants
gender_sport_participation = merged_df.groupby(['Gender', 'Sport']).size().reset_index(name='Participants')

# Sort by Gender and Participants
gender_sport_participation = gender_sport_participation.sort_values(['Gender', 'Participants'], ascending=[True, False]).reset_index(drop=True)

# Display top 10 sports per gender
top_sports_by_gender = gender_sport_participation.groupby('Gender').head(10).reset_index(drop=True)

print(top_sports_by_gender)
```

```python
sns.barplot(data=top_sports_by_gender, x='Sport', y='Participants', hue='Gender')

# Rotate x-axis labels for better readability
plt.xticks(rotation=45)
plt.title('Top 5 Sports by Gender-wise Participation')
plt.ylabel('Number of Participants')
plt.xlabel('Sport')
plt.show()
```
